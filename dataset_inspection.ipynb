```python
csv_files = []

for run in labels.keys():
    csv_files.extend(list((data_path/run).rglob("*.csv")))
    
len(csv_files)
```

```python
test_2_channels = []
test_3_channels = []

for csv_file in csv_files:
    if len(pd.read_csv(csv_file).columns) == 3:
        test_2_channels.append(csv_file)
    elif len(pd.read_csv(csv_file).columns) == 4:
        test_3_channels.append(csv_file)
        
len(test_2_channels), len(test_3_channels)
```
